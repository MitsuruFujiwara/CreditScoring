SVMによる多クラス分類

　　必要なパッケージをインポート

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, accuracy_score
```

データをインポート

```python
df = pd.read_csv('TrainingData_rev.csv').fillna(0)
df_test = pd.read_csv('testData_rev.csv').fillna(0)
```

```python
df
```

```python
df_test
```

XとYに分割

```python
df_y = df['Ratings #']
df_x = df.drop('Ratings #', 1)
df_test_y = df_test['Ratings #']
df_test_x = df_test.drop('Ratings #', 1)
```

SVMで分析

```python
svm = SVC()
svm.fit(df_x, df_y)
```

結果を集計

```python
train_pred = svm.predict(df_x)
test_pred = svm.predict(df_test_x)
```

学習用データに対する正解率

```python
accuracy_score(df_y, train_pred)
```

テスト用データに対する正解率

```python
accuracy_score(df_test_y, test_pred)
```
