必要なパッケージをインポート

```python
import pandas as pd
from sklearn.decomposition import PCA
from SetData import ConvertData
```

元データと出力先のパスを設定

```python
datapath = 'TrainingData_raw.csv'
outputpath = 'TrainingData.csv'
```

データを標準化

```python
c = ConvertData(datapath, outputpath, 'Ratings #')
df = c.main().fillna(0)
df_y = df['Ratings #']
df_x = df.drop('Ratings #', 1)
```

```python
pca = PCA(4)
pca.fit(df_x, df_y)
```

```python
res = pca.explained_variance_ratio_
```

```python
res.sum()
```

```python
df_new = pca.transform(df_x, df_y)
```

```python
df_new
```

```python
def col(x):
    for i, t in enumerate(x):
        yield 'X' + str(i+1)
```

```python
_col = list(col(df_new[0]))
```

```python
df_new = pd.DataFrame(df_new, columns=_col)
```

```python
df_new
```

```python
df_new ['Y'] = df_y
```
