scikit-learnの機能を使ってデータ分割

```python
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
from SetData import ConvertData
```

```python
cd_tr = ConvertData('TrainingData_raw_rev.csv', 'TrainingData_rev.csv', 'Ratings #')
df = cd_tr.main()
```

```python
df
```

XとYに分割

```python
df_x, df_y = df.drop('Ratings #', 1).fillna(0), df['Ratings #']
```

トレーニングデータセット（70％）とテストデータセット（30％）に分割

```python
trX, testX, trY, testY = train_test_split(df_x, df_y, test_size=0.3, random_state=0)
```
